```python
import glob
import numpy as np
import sys
sys.path.append("..")
import utils as ut
```

```python
directoryPath = "perm_techn_2/"
files = []
for file in glob.glob(directoryPath+"*.csv"):
    x = np.genfromtxt(file,delimiter=',')
    files.append(x)
```

```python
corr_spectral = [item[0] for item in files]
corr_polo = [item[1] for item in files]
corr_polo_flipped = [item[2] for item in files]
corr_polo_flipped_rows = [item[3] for item in files]
corr_mix_1 = [item[4] for item in files]
corr_mix_2 = [item[5] for item in files]
corr_random_rows = [item[6] for item in files]
corr_random_columns = [item[7] for item in files]
corr_mean_rows_incr = [item[8] for item in files]
corr_min_rows_incr = [item[9] for item in files]
corr_max_rows_incr = [item[10] for item in files]
corr_mean_rows_decr = [item[11] for item in files]
corr_min_rows_decr = [item[12] for item in files]
corr_max_rows_decr = [item[13] for item in files]
corr_mean_columns_incr = [item[14] for item in files]
corr_min_columns_incr = [item[15] for item in files]
corr_max_columns_incr = [item[16] for item in files]
corr_mean_columns_decr = [item[17] for item in files]
corr_min_columns_decr = [item[18] for item in files]
corr_max_columns_decr = [item[19] for item in files]
```

```python
# all on one plot
data = [corr_spectral, corr_polo, corr_polo_flipped, corr_polo_flipped_rows, corr_mix_1, corr_mix_2, corr_random_rows, corr_random_columns,
       corr_mean_rows_incr, corr_min_rows_incr, corr_max_rows_incr, corr_mean_rows_decr, corr_min_rows_decr, corr_max_rows_decr,
       corr_mean_columns_incr, corr_min_columns_incr, corr_max_columns_incr, corr_mean_columns_decr, corr_min_columns_decr, corr_max_columns_decr]
data = np.array(data)
temp = np.median(data, axis=1).argsort()
data = data[temp]
data = data.tolist()
labels = ['spectral', 'polo', 'polo flipped', 'polo flipped rows','rows min INCR + columns max INCR', 'rows min DECR + columns max INCR', 'random rows', 'random columns',
         'rows mean INCR', 'rows min INCR', 'rows max INCR', 'rows mean DECR', 'rows min DECR', 'rows max DECR',
         'columns mean INCR', 'columns min INCR', 'columns max INCR', 'columns mean DECR', 'columns min DECR', 'columns max DECR']
labels = np.array(labels)
labels = labels[temp]
labels = labels.tolist()
ut.violin_plot_all(data, labels, '', "violin_plots_analysis_2.png")
```
