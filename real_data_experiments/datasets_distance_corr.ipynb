```python
import numpy as np
from sklearn import cluster
import matplotlib.pyplot as plt
import copy
```

```python
def plot_eight_datasets_distance_corr(data1, title1, data1_orig, data2, title2, data2_orig, data3, title3, data3_orig, 
                        data4, title4, data4_orig, data5, title5, data5_orig, data6, title6, data6_orig,
                        data7, title7, data7_orig, data8, title8, data8_orig,):
    
    data1, data2, data3, data4, data5, data6, data7, data8 = data1.ravel(), data2.ravel(), data3.ravel(), data4.ravel(), data5.ravel(), data6.ravel(), data7.ravel(), data8.ravel()
    data1_orig, data2_orig, data3_orig, data4_orig, data5_orig, data6_orig, data7_orig, data8_orig = data1_orig.ravel(), data2_orig.ravel(), data3_orig.ravel(), data4_orig.ravel(), data5_orig.ravel(), data6_orig.ravel(), data7_orig.ravel(), data8_orig.ravel()
    
    n_bins = 20 #"auto"
    x_label = "Euclidean norm"
    y_label = "density"
    
    fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4, figsize=(16, 8))
    
    ax1.hist(data1, density=True, bins=n_bins, label="STMF", alpha=0.5)
    ax1.hist(data1_orig, density=True, bins=n_bins, label="NMF", alpha=0.5, color="red")
    ax1.set_title(title1)
    ax1.set_xlabel(x_label)
    ax1.set_ylabel(y_label)
    ax1.legend()

    ax2.hist(data2, density=True, bins=n_bins, label="STMF", alpha=0.5)
    ax2.hist(data2_orig, density=True, bins=n_bins, label="NMF", alpha=0.5, color="red")
    ax2.set_title(title2)
    ax2.set_xlabel(x_label)
    ax2.set_ylabel(y_label)
    ax2.legend()
    
    ax3.hist(data3, density=True, bins=n_bins, label="STMF", alpha=0.5)
    ax3.hist(data3_orig, density=True, bins=n_bins, label="NMF", alpha=0.5, color="red")
    ax3.set_title(title3)
    ax3.set_xlabel(x_label)
    ax3.set_ylabel(y_label)
    ax3.legend()
    
    ax4.hist(data4, density=True, bins=n_bins, label="STMF", alpha=0.5)
    ax4.hist(data4_orig, density=True, bins=n_bins, label="NMF", alpha=0.5, color="red")
    ax4.set_title(title4)
    ax4.set_xlabel(x_label)
    ax4.set_ylabel(y_label)
    ax4.legend()
    
    ax5.hist(data5, density=True, bins=n_bins, label="STMF", alpha=0.5)
    ax5.hist(data5_orig, density=True, bins=n_bins, label="NMF", alpha=0.5, color="red")
    ax5.set_title(title5)
    ax5.set_xlabel(x_label)
    ax5.set_ylabel(y_label)
    ax5.legend()
    
    ax6.hist(data6, density=True, bins=n_bins, label="STMF", alpha=0.5)
    ax6.hist(data6_orig, density=True, bins=n_bins, label="NMF", alpha=0.5, color="red")
    ax6.set_title(title6)
    ax6.set_xlabel(x_label)
    ax6.set_ylabel(y_label)
    ax6.legend()
    
    ax7.hist(data7, density=True, bins=n_bins, label="STMF", alpha=0.5)
    ax7.hist(data7_orig, density=True, bins=n_bins, label="NMF", alpha=0.5, color="red")
    ax7.set_title(title7)
    ax7.set_xlabel(x_label)
    ax7.set_ylabel(y_label)
    ax7.legend()
    
    ax8.hist(data8, density=True, bins=n_bins, label="STMF", alpha=0.5)
    ax8.hist(data8_orig, density=True, bins=n_bins, label="NMF", alpha=0.5, color="red")
    ax8.set_title(title8)
    ax8.set_xlabel(x_label)
    ax8.set_ylabel(y_label)
    ax8.legend()

    fig.tight_layout()
    fig.subplots_adjust(top=0.82)
    fig.subplots_adjust(wspace=0.3)
    fig.savefig('datasets_distance_corr.png', bbox_inches='tight', dpi=300);
```
