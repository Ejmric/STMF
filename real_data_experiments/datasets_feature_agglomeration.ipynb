```python
import numpy as np
from sklearn import cluster
import matplotlib.pyplot as plt
import copy
```

```python
def feat_agg(location):
    # importing data
    original_data = np.genfromtxt(location, delimiter='\t').T
    original_data = np.delete(original_data, 0, axis=0)
    original_data = np.delete(original_data, 0, axis=1)
    orig_data = copy.deepcopy(original_data)
    # feature agglomeration columns
    agglo = cluster.FeatureAgglomeration(n_clusters=100)
    agglo.fit(original_data)
    transf_data = agglo.transform(original_data)
    return orig_data, transf_data
```

```python
def import_data(location):
    # importing data
    original_data = np.genfromtxt(location, delimiter='\t').T
    original_data = np.delete(original_data, 0, axis=0)
    original_data = np.delete(original_data, 0, axis=1)
    return original_data
```

```python
def plot_data(data, title, data_orig):
    data = data.ravel()
    data_orig = data_orig.ravel()
    plt.hist(data, density=True, bins=10, label="Agglomeration", alpha=0.5)
    plt.hist(data_orig, density=True, bins=10, label="Original", alpha=0.5)
    plt.title(title)
    plt.legend();
```

```python
def plot_eight_datasets(data1, title1, data1_orig, data2, title2, data2_orig, data3, title3, data3_orig, 
                        data4, title4, data4_orig, data5, title5, data5_orig, data6, title6, data6_orig,
                        data7, title7, data7_orig, data8, title8, data8_orig,):
    
    data1, data2, data3, data4, data5, data6, data7, data8 = data1.ravel(), data2.ravel(), data3.ravel(), data4.ravel(), data5.ravel(), data6.ravel(), data7.ravel(), data8.ravel()
    data1_orig, data2_orig, data3_orig, data4_orig, data5_orig, data6_orig, data7_orig, data8_orig = data1_orig.ravel(), data2_orig.ravel(), data3_orig.ravel(), data4_orig.ravel(), data5_orig.ravel(), data6_orig.ravel(), data7_orig.ravel(), data8_orig.ravel()
    
    fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4, figsize=(16, 8))
    
    ax1.hist(data1, density=True, bins=10, label="Agglomeration", alpha=0.5)
    ax1.hist(data1_orig, density=True, bins=10, label="Original", alpha=0.5)
    ax1.set_title(title1)
    ax1.legend()

    ax2.hist(data2, density=True, bins=10, label="Agglomeration", alpha=0.5)
    ax2.hist(data2_orig, density=True, bins=10, label="Original", alpha=0.5)
    ax2.set_title(title2)
    ax2.legend()
    
    ax3.hist(data3, density=True, bins=10, label="Agglomeration", alpha=0.5)
    ax3.hist(data3_orig, density=True, bins=10, label="Original", alpha=0.5)
    ax3.set_title(title3)
    ax3.legend()
    
    ax4.hist(data4, density=True, bins=10, label="Agglomeration", alpha=0.5)
    ax4.hist(data4_orig, density=True, bins=10, label="Original", alpha=0.5)
    ax4.set_title(title4)
    ax4.legend()
    
    ax5.hist(data5, density=True, bins=10, label="Agglomeration", alpha=0.5)
    ax5.hist(data5_orig, density=True, bins=10, label="Original", alpha=0.5)
    ax5.set_title(title5)
    ax5.legend()
    
    ax6.hist(data6, density=True, bins=10, label="Agglomeration", alpha=0.5)
    ax6.hist(data6_orig, density=True, bins=10, label="Original", alpha=0.5)
    ax6.set_title(title6)
    ax6.legend()
    
    ax7.hist(data7, density=True, bins=10, label="Agglomeration", alpha=0.5)
    ax7.hist(data7_orig, density=True, bins=10, label="Original", alpha=0.5)
    ax7.set_title(title7)
    ax7.legend()
    
    ax8.hist(data8, density=True, bins=10, label="Agglomeration", alpha=0.5)
    ax8.hist(data8_orig, density=True, bins=10, label="Original", alpha=0.5)
    ax8.set_title(title8)
    ax8.legend()

    fig.tight_layout()
    fig.subplots_adjust(top=0.82)
    fig.subplots_adjust(wspace=0.3)
    fig.savefig('datasets.png', bbox_inches='tight', dpi=300);
```

```python
# data
AML_orig, AML = feat_agg('../multi_omic/AML/AML_gene_subset.txt')
COLON_orig, COLON = feat_agg('../multi_omic/Colon/Colon_gene_subset.txt')
GBM_orig, GBM = feat_agg('../multi_omic/GBM/GBM_gene_subset.txt')
LIHC_orig, LIHC = feat_agg('../multi_omic/LIHC/LIHC_gene_subset.txt')
LUSC_orig, LUSC = feat_agg('../multi_omic/LUSC/LUSC_gene_subset.txt')
OV_orig, OV = feat_agg('../multi_omic/OV/OV_gene_subset.txt')
SARC_orig, SARC = feat_agg('../multi_omic/SARC/SARC_gene_subset.txt')
SKCM_orig, SKCM = feat_agg('../multi_omic/SKCM/SKCM_gene_subset.txt')
```

```python
plot_eight_datasets(AML, "AML", AML_orig, COLON, "COLON", COLON_orig, GBM, "GBM", GBM_orig, LIHC, "LIHC", LIHC_orig, LUSC, "LUSC", LUSC_orig, OV, "OV", OV_orig, SKCM, "SKCM", SKCM_orig, SARC, "SARC", SARC_orig)
```
